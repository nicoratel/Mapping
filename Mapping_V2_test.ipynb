# Extraction des profils azimutahux: 

Remarque: 

L'utilisation du Mapping contraint à avoir des valeurs de $\beta$ comprises entre 0 et 180°. On ne peut pas avoir accès au profil azimuthal complet, bien qu'il existe (en réalité il en manque une partie, mais on peut avoir le pic centré sur $\beta=0 ou 180°$ si on arrive à extraire le profil azimuthal sour toute la plage accessible).

Afin d'y remédier, les profils azimuthaux SAXS et USAXS ont été extraits avec pyFAI (integrate_radial). Les corrections apportées par la projection des data dans l'espace réciproque sont négligeables aux petits et très petits angles, ce qui rend cette approche légitime.

Dans ce qui suit, seuls les profils WAXS ont été extraits après projection des données dans l'espace réciproque. La comparaison des 2 méthodes (pyFAI et mapping) dans les domaines SAXS et USAXS donne des résultats très similaires (non présentés ici)
